```python
import numpy as np

# DATOS PERSONALIZADOS (cada alumno cambia estos valores)
mesada_semanal = 500  # Mesada que recibes
gastos = np.array([80, 45, 120, 60, 90, 30, 75])  # Gastos diarios

# CÃLCULOS
total_gastado = np.sum(gastos)
promedio_diario = np.mean(gastos)
dia_mayor_gasto = np.argmax(gastos) + 1
saldo = mesada_semanal - total_gastado

# RESULTADOS
print(f"ðŸ’° MI MESADA SEMANAL: ${mesada_semanal}")
print(f"ðŸ“Š Total gastado: ${total_gastado}")
print(f"ðŸ“ˆ Promedio diario: ${promedio_diario:.2f}")
print(f"ðŸ›ï¸ DÃ­a de mayor gasto: DÃ­a {dia_mayor_gasto}")
print(f"ðŸ’µ Saldo final: ${saldo}")

if saldo > 0:
    print("âœ… Â¡Buen manejo de dinero!")
else:
    print("âŒ Â¡Cuidado con los gastos!")
```

```python
import numpy as np

# DATOS PERSONALIZADOS
hobbies = np.array(['Videojuegos', 'Deporte', 'MÃºsica', 'Lectura', 'Redes'])
tiempo_semanal = np.array([10, 6, 8, 5, 12])  # Horas por semana

# CÃLCULOS
tiempo_total = np.sum(tiempo_semanal)
hobby_mas_tiempo = hobbies[np.argmax(tiempo_semanal)]
porcentajes = (tiempo_semanal / tiempo_total) * 100
tiempo_promedio = np.mean(tiempo_semanal)

# RESULTADOS
print("ðŸŽ¯ ANÃLISIS DE MIS HOBBIES")
print(f"â° Tiempo total semanal: {tiempo_total} horas")
print(f"ðŸ† Hobby favorito: {hobby_mas_tiempo}")
print(f"ðŸ“Š Tiempo promedio por hobby: {tiempo_promedio:.1f} horas")
print("\nðŸ“ˆ DistribuciÃ³n porcentual:")
for i, hobby in enumerate(hobbies):
    print(f"   {hobby}: {porcentajes[i]:.1f}%")
```

```python
import pandas as pd

# DATOS PERSONALIZADOS
datos_academicos = {
    'materia': ['MatemÃ¡ticas', 'ProgramaciÃ³n', 'Historia', 'InglÃ©s', 'Ciencias'],
    'primer_parcial': [85, 92, 78, 88, 82],
    'segundo_parcial': [90, 95, 82, 85, 79],
    'proyecto_final': [88, 98, 75, 90, 80],
    'creditos': [4, 5, 3, 2, 4]
}

df = pd.DataFrame(datos_academicos)

# CÃLCULOS
df['promedio_final'] = (df['primer_parcial'] + df['segundo_parcial'] + df['proyecto_final']) / 3
df['puntos_honor'] = df['promedio_final'] * df['creditos']

promedio_general = df['promedio_final'].mean()
mejor_materia = df.loc[df['promedio_final'].idxmax(), 'materia']
peor_materia = df.loc[df['promedio_final'].idxmin(), 'materia']

# RESULTADOS
print("ðŸ“š MI RENDIMIENTO ACADÃ‰MICO")
print("\nðŸ“Š Calificaciones finales:")
print(df[['materia', 'promedio_final']].round(2))

print(f"\nðŸŽ¯ ESTADÃSTICAS:")
print(f"Promedio general: {promedio_general:.2f}")
print(f"Mejor materia: {mejor_materia}")
print(f"Materia a mejorar: {peor_materia}")

print("\nðŸ“ˆ EvoluciÃ³n por materia:")
for _, materia in df.iterrows():
    progreso = materia['proyecto_final'] - materia['primer_parcial']
    tendencia = "ðŸ“ˆ" if progreso > 0 else "ðŸ“‰" if progreso < 0 else "âž¡ï¸"
    print(f"   {materia['materia']}: {progreso:+d} puntos {tendencia}")
```

```python
import matplotlib.pyplot as plt

# DATOS PERSONALIZADOS
actividades = ['Dormir', 'Estudiar', 'Trabajo', 'Hobbies', 'Comida', 'Transporte', 'Redes']
horas = [8, 6, 4, 2, 2, 1, 1]
colores = ['#FF9999', '#66B2FF', '#99FF99', '#FFD700', '#FFB6C1', '#C9A0DC', '#FFA07A']

# CONFIGURACIÃ“N
plt.figure(figsize=(10, 8))

# DIAGRAMA CIRCULAR
plt.subplot(2, 1, 1)
wedges, texts, autotexts = plt.pie(horas, labels=actividades, colors=colores,
                                   autopct='%1.1f%%', startangle=90)
plt.title('ðŸ•’ Mi DistribuciÃ³n de Tiempo Diario', fontsize=14, fontweight='bold')

# Mejorar estÃ©tica
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# GRÃFICO DE BARRAS HORIZONTAL
plt.subplot(2, 1, 2)
plt.barh(actividades, horas, color=colores, edgecolor='black')
plt.xlabel('Horas')
plt.title('ðŸ“Š Tiempo por Actividad (Horas)')
plt.grid(axis='x', alpha=0.3)

# AÃ±adir valores en las barras
for i, v in enumerate(horas):
    plt.text(v + 0.1, i, f'{v}h', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

# ANÃLISIS
print("â° ANÃLISIS DE MI DÃA:")
print(f"Total horas: {sum(horas)}")
print(f"Actividad principal: {actividades[horas.index(max(horas))]} ({max(horas)} horas)")
print(f"Porcentaje de estudio/trabajo: {((horas[1] + horas[2])/sum(horas)*100):.1f}%")
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# DATOS PERSONALIZADOS
datos_ejercicio = {
    'dia': ['Lun', 'Mar', 'MiÃ©', 'Jue', 'Vie', 'SÃ¡b', 'Dom'] * 4,
    'semana': ['Sem1']*7 + ['Sem2']*7 + ['Sem3']*7 + ['Sem4']*7,
    'tipo_ejercicio': ['Cardio', 'Fuerza', 'Descanso', 'Cardio', 'Fuerza', 'Flexibilidad', 'Descanso'] * 4,
    'minutos': [45, 60, 0, 30, 75, 20, 0, 50, 65, 0, 40, 70, 25, 0,
                60, 55, 0, 35, 80, 30, 0, 40, 70, 0, 45, 75, 15, 0],
    'calorias_quemadas': [320, 450, 0, 280, 520, 150, 0, 350, 480, 0, 310, 500, 180, 0,
                         420, 400, 0, 290, 550, 220, 0, 310, 490, 0, 330, 520, 120, 0]
}

df = pd.DataFrame(datos_ejercicio)

# CONFIGURACIÃ“N
sns.set_style("whitegrid")
plt.figure(figsize=(15, 10))

# GRÃFICO 1: EvoluciÃ³n semanal
plt.subplot(2, 2, 1)
sns.lineplot(data=df, x='dia', y='minutos', hue='semana', marker='o', linewidth=2.5)
plt.title('ðŸƒ EvoluciÃ³n de Minutos de Ejercicio por Semana')
plt.legend(title='Semana')

# GRÃFICO 2: ComparaciÃ³n por tipo de ejercicio
plt.subplot(2, 2, 2)
sns.boxplot(data=df[df['minutos'] > 0], x='tipo_ejercicio', y='minutos', palette='Set2')
plt.title('ðŸ“Š DistribuciÃ³n de Tiempo por Tipo de Ejercicio')
plt.xticks(rotation=45)

# GRÃFICO 3: CalorÃ­as vs Minutos
plt.subplot(2, 2, 3)
sns.scatterplot(data=df[df['minutos'] > 0], x='minutos', y='calorias_quemadas',
                hue='tipo_ejercicio', s=100, palette='viridis')
plt.title('ðŸ”¥ RelaciÃ³n: Minutos vs CalorÃ­as Quemadas')

# GRÃFICO 4: Heatmap de actividad
plt.subplot(2, 2, 4)
pivot_data = df.pivot_table(values='minutos', index='semana', columns='dia', aggfunc='sum')
sns.heatmap(pivot_data, annot=True, cmap='YlOrRd', linewidths=1)
plt.title('ðŸ“… Heatmap de Actividad Semanal')

plt.tight_layout()
plt.show()

# RESUMEN ESTADÃSTICO
print("ðŸ’ª RESUMEN DE MI ENTRENAMIENTO:")
print(f"Total minutos: {df['minutos'].sum()}")
print(f"Promedio diario: {df[df['minutos'] > 0]['minutos'].mean():.1f} minutos")
print(f"Tipo de ejercicio favorito: {df.groupby('tipo_ejercicio')['minutos'].sum().idxmax()}")
```
